```python
#Plot for milage across year
plt.figure(figsize = (21,5))
sns.boxplot(x = 'year', y = 'mileage', data = df, palette = 'Set3')
plt.title ('Mileage across years')
plt.xlabel ('Year')
plt.ylabel ('Mileage')
plt.show()
```

```python
#Plot for price across year
plt.figure(figsize = (25,8))
sns.boxplot(x = 'year', y = 'price', data = df, palette = 'Set3')
plt.title ('price across years')
plt.xlabel ('Year')
plt.ylabel ('Price')
plt.show()
```

```python
#Pie Chart for Fuel_D
plt.figure(figsize =(5,5))
plt.pie(df['fuel_ID'].value_counts().values,
        labels = df['fuel_ID'].value_counts().index,
        autopct = '%.2f%%', startangle = 140)
plt.title('Fuel Distribution')
plt.show()
```

```python
#Pie Chart for transmission_ID denoting the composition
plt.figure(figsize =(5,5))
plt.pie(df['transmission_ID'].value_counts().values,
        labels = df['transmission_ID'].value_counts().index,
        autopct = '%.2f%%', startangle = 140)
plt.title('transmission_ID denoting the composition')
plt.show()
```

```python
# correlation Analysis
cm = df. corr()
plt.figure(figsize = (9,5))
sns.heatmap(cm, annot = True)
plt.show()
```
